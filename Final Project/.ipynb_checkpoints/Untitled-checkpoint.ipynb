```python
import requests
import pandas as pd
import os
from dotenv import load_dotenv
import json
from json import dumps, loads
from pandas import json_normalize
```

```python
url = 'https://cryptingup.com/api/'
```

```python
r = requests.get('https://cryptingup.com/api/assets?start=8423129941')
data = r.json()
```

```python
print(json.dumps(data, indent=2))
```

```python
r.status_code
```

```python
print(type(data))
print(data.keys())
```

```python

data_df = json_normalize(data['assets'])
data_df.head(10)
```

```python
data_df.describe()
```

```python
exchange = requests.get('https://cryptingup.com/api/exchanges')
data_exchange = exchange.json()
```

```python
print(json.dumps(data_exchange, indent=2))
```

```python
print(type(data_exchange))
print(data_exchange.keys())
```

```python
data_exchange_df = json_normalize(data_exchange['exchanges'])
data_exchange_df.head()
data_exchange_df.sort_values(by = ['exchange_id'])
```

```python
data_exchange_df.describe()
```

```python
markets = requests.get('https://cryptingup.com/api/exchanges/BINANCE/markets')
data_markets = markets.json()
print(json.dumps(data_markets, indent=2))
```

```python
print(type(data_markets))
print(data_markets.keys())
```

```python
data_markets_df = json_normalize(data_markets['markets'])
data_markets_df.head()
data_markets_df.sort_values(by = ['symbol'])
```
